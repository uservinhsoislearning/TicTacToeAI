```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = []
data.append(pd.read_csv('MiniMax.csv'))
data.append(pd.read_csv('AlphaBeta.csv'))
data.append(pd.read_csv('MCTS.csv'))

data
```

```python
first_move = []
for i in range(3):
    tmp = data[i]
    first_move.append(tmp.loc[tmp['Move No.'] == 1])
```

```python
first_move_total = pd.concat([first_move[0], first_move[1], first_move[2]])
```

```python
first_move_total
```

```python
second_move = []
for i in range(3):
    tmp = data[i]
    second_move.append(tmp.loc[tmp['Move No.'] == 2])
```

```python
second_move_total = pd.concat([second_move[0],second_move[1],second_move[2]])
```

```python
second_move_total
```

```python
total_move = pd.concat([first_move_total, second_move_total])
total_move
```

```python
plt.figure(figsize=(12,6))

plt.plot('Play', 'Time per move', data=first_move[0], label='MiniMax')
plt.plot('Play', 'Time per move', data=first_move[1], label='Alpha-Beta pruning')
plt.plot('Play', 'Time per move', data=first_move[2], label='Monte Carlo Tree Search')

plt.xlabel('Plays')
plt.ylabel('Time per move (ms)')

plt.xticks(first_move[0]['Play'])

plt.title('Time per move by plays')
plt.legend()

plt.show()
```
